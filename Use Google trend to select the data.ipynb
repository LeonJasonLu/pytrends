## 定義一個讀入一個檔案，每次可以向google trend查詢40個詞彙的函式(check_and_write)，並將查詢後符合結果的詞彙輸出至"output.txt" 
## 如果能利用代理伺服器於每次查詢完40個字後換ip的話，理論上可以加快處理完所有資料的速度。但是我現在尚沒辦法找到實行的辦法

```python
from pytrends.request import TrendReq
import time

def check_and_write(inputFile):
    ##調用pytrends並設定本地端的時區以及語言
    pytrend = TrendReq(hl = 'zh-TW', tz = 360, timeout=(10,25)) ## hl(host language) & tz(time zone)
    wordsList = []  ##用來存放檔案中的所有詞彙
    with open(inputFile, "r", encoding = 'utf8') as inputfile:
        for word in inputfile:
            word = word.strip('\n')  ##去除每行詞彙於讀入檔案時都會被加上的\n 
            wordsList.append(word)            

    for i in range(0, len(wordsList), 40):
        print(int((i+40)/40))
        partialList = wordsList[i : i + 40]
        checkedResultList = []  ##每輪查詢的40詞彙中的合格詞彙會被暫時放置於此，直到開始查下一輪的40個詞彙
        for j in range(0, len(partialList)):
            queryList = [""] ##要放進pytrend裡的參數，每次會取partialList中的一個詞彙進行查詢
            queryList[0] = partialList[j]
    
            pytrend.build_payload(
            kw_list = queryList,
            cat = 0,  ## cat(category)為零代表查詢google trend的所有類別。若cat=14代表專門查詢"人文社會"類別中的資料
            timeframe = 'today 3-m', ## 搜尋自3個月前至今的每日資料
            geo = 'TW',  ## geo(geography)，TW是台灣代碼
            gprop = '')  ## gprop(Google property)，搜尋結果的類型，有image, news, youtube...
    
            dataframe = pytrend.interest_over_time()  ## 取得數據
    
##先判定該詞彙是否能再google trend被查詢到。若可，pytrends會回傳一個非空的dataframe；若否，pytrends會回傳一個空dataframe
            if(dataframe.empty == False): 
                sumOfThisWord = (dataframe.sum()).values[0]  ##計算該詞彙在三個月內總共出現幾次
                if sumOfThisWord >= 100:  ##如果至少出現超過100次，即可將該詞彙加入checkedResultlList
                    checkedResultList.append(partialList[j])

        print(checkedResultList)
        for word in checkedResultList:
            print(word, end ='\n', file=open('output.txt','a+', encoding = "utf8")) 
        
        if (i != len(wordsList)-40):
            print("先睡一下(60秒)")
            ##如果短時間內同一ip向google trend送出太多查詢請求的話，google trend會暫停服務該ip一段未知的時間(ErrorCode:429)  
            time.sleep(60)  
    print("Congratulations!!! This file is completely processed!!!")
    
```
